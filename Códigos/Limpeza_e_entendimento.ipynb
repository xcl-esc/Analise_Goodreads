```python
import pandas as pd
import numpy as np
import os
import pyarrow
```

```python
path = '/content/drive/MyDrive/projeto_novo/Dados/books.csv'
df = pd.read_csv(path, on_bad_lines='skip')
df.info()
```

```python
df.head()
```

Problemas encontrados:

- O rótulo 'num_pages' contém um espaço vazio antes do nome
- rotulo isbn formatado como texto (pq?) - O rótulo isbn é uma identificação do livro, por isso está setado como texto(como um id)
- publication_date precisa passar para date_time
- livros com 0 avaliações mas possuem notas (possível erro de extração, ou correlações errôneas com edições diferentes do mesmo livro entre outros problemas). Não há possibilidade de correção por possuírem vários casos diferentes, logo ficarão como estão.

```python
df.columns = df.columns.str.strip()
```

```python
df.info()
```

```python
df.tail()
```

```python
#checagem pré-conversão
print(df.loc[[8177, 11094], ['publication_date', 'title']])
```

```python
df['publication_date'] = pd.to_datetime(df['publication_date'], errors='coerce')
```

```python
df.info()
```

```python
df['publication_date'].isna().sum()
```

Precisamos ver quem são as linhas com datas nulas pra decidir se devem ser excluídas ou não.

```python
checagem = df[df['publication_date'].isna()]
checagem
```

Essas duas linhas:
- 8177        11/31/2000  In Pursuit of the Proper Sinner (Inspector Lyn...)
- 11094        6/31/1982         Montaillou  village occitan de 1294 à 1324

Possuem erros de preenchimento de datas, no primeiro e no segundo caso (nos meses 11 e 6) o mes possui apenas 30 dias, e em ambos está preenchido com dia ou mes errados. Por esse motivo não há como tratar essas linhas por não haver informações que respaldem qual o erro com precisão e essas linhas serão excluídas do dataset. O impacto é quase nenhum considerando que representam 0,017% do dataset.

```python
df.dropna(subset=['publication_date'], inplace=True)
df.info()
```
